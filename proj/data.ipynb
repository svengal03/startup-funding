```python
df.drop(['Remarks'],axis=1)
```

```python
df['AmountInUSD'] = df['AmountInUSD'].apply(lambda x:float(str(x).replace(",","")))
```

```python
df['Date']=df['Date'].replace({"12/05.2015":"12/05/2015"})
df['Date']=df['Date'].replace({"13/04.2015":"13/04/2015"})
df['Date']=df['Date'].replace({"15/01.2015":"15/01/2015"})
df['Date']=df['Date'].replace({"22/01//2015":"22/01/2015"})
df
```

```python
df["ipm"] = (pd.to_datetime(df['Date'],format='%d/%m/%Y').dt.month)+(pd.to_datetime(df['Date'],format='%d/%m/%Y').dt.year*100)
temp = df['ipm'].value_counts().sort_values(ascending = False)
print("Number of funding per month in decreasing order (Funding Wise)\n\n",temp)
year_month = df['ipm'].value_counts()
```

```python
plt.figure(figsize=(15,8))
sns.barplot( df.ipm,df.AmountInUSD, alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Amonut Of Investments', fontsize=12)
plt.title("YearMonth - Number of fundings distribution", fontsize=16)
plt.show()
```

```python
fundcount  = df['StartupName'].value_counts()
fundcount = fundcount.head(20)
print(fundcount)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(fundcount.index, fundcount.values, alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('Startups', fontsize=12)
plt.ylabel('Number of fundings made', fontsize=12)
plt.title("Startups-Number of fundings distribution", fontsize=16)
plt.show()
```
